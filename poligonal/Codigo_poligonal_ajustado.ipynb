```python
import numpy as np
import pandas as pd
```

```python
# Función que convierte los datos de azimut en ángulos euclidianos

def az2carto(azi):
  return -azi + 90
```

**Ajuste del código con variables conocidas**

```python
#Ejemplo A: segmento de control
# Segmento de 50 m, perpendicular al rumbo de la capa x (paralelo a la dirección de buzamiento) y con ángulo de inclinación 0 con respecto a la horizontal
## Variables ingresadas
L=(50); # Longitud del segmento
I=45*np.pi/180; # Ángulo de inclinación de la capa
a=az2carto(0)*np.pi/180; # Dirección de la capa en azimut del rumbo
b=az2carto(90)*np.pi/180; # Dirección de la inclinacion de la capa en azimut del buzamiento
f=az2carto(90)*np.pi/180; # Dirección del segmento en azimut del rumbo
s=0*np.pi/180; # Ángulo de inclinación del segmento
```

```python
#Cálculo de variables reales
L2=abs(np.cos((s))*L) # Longitud L corregida con respecto a su ángulo de inclinación
I2=abs(np.arctan(np.tan(I)/np.cos(abs(b-f)))) # Ángulo de inclinación real
E=np.sin(I2)*L2 # Espesor real para el ejemplo de las variables conocidas
```

```python
# Definir función para el cálculo del espesor real

def cal_thick(S,L,I,a,b,f):
  '''
  cal_thick: Calcular el espesor real a partir de los datos medidos en una poligonal
  Hecho por: Paul Goyes (goyes.yesid@gmail.com) y Ana Mantilla (anagmd2019@gmail.com)

  Variables de entrada
  S: Ángulo de inclinación del segmento
  L: Longitud de la vía o segmento a calcular espesor aparente (metros)
  I: Inclinación de la capa
  a: Dirección de la capa en azi del rumbo
  b: Dirección de la inclinación de la capa en azi del buzamiento
  f: Dirección del segmento en azi del rumbo'''
  ######## convertir de az2carto
  I=I*np.pi/180; #Ángulo de inclinación de la capa
  S=S*np.pi/180; # Ángulo de inclinación del segmento
  a=az2carto(a)*np.pi/180; # Dirección de la capa en azimut del rumbo
  b=az2carto(b)*np.pi/180; # Dirección de la inclinacion en azimut del buzamiento
  f=az2carto(f)*np.pi/180; # Dirección del segmento en azimut del rumbo
  ######## Calcular variables reales
  L_real= abs(np.cos((S))*L) # Longitud L corregida con respecto al ángulo de inclinación
  I_real= abs(np.arctan(np.tan(I)/np.cos(abs(b-f)))) # Ángulo de inclinación real de la capa
  E_real= np.sin(I_real)*L_real # Espesor real para el ejemplo de ajuste
  return E_real
```
